PC

```python
pca = PCA()
Z = pca.fit_transform(X[numeric_columns])

# (High)->1,  (Low)->0

idx_High= np.where(y == 0)
idx_Low = np.where(y == 1)


plt. figure(figsize=(15,10))
plt.scatter(Z[idx_High,0], Z[idx_High,1], c='g', label='HIGH')
plt.scatter(Z[idx_Low,0], Z[idx_Low,1], c='r', label='LOW')
```

```python
A = pca.components_.T
plt. figure(figsize=(8,4))

plt.scatter(A[:,0],A[:,1],c='r')
plt.xlabel('A1')
plt.ylabel('A2')
for label, x, y in zip(variables, A[:, 0], A[:, 1]):
  plt.annotate(label, xy=(x, y), xytext=(-2, 2), textcoords='offset points', ha='right', va='bottom')
```

SCREE PLOT

```python
#Eigenvalues
Lambda = pca.explained_variance_

#Scree plot
x = np.arange(len(Lambda)) + 1
plt. figure(figsize=(15,10))
plt.plot(x,Lambda/sum(Lambda), 'ro-', lw=3)
plt.xticks(x, [""+str(i) for i in x], rotation=0)
plt.xlabel('Number of components')
plt.ylabel('Explained variance')
Lambda
```

EXPLAINED VARIANCE

```python
ell = pca.explained_variance_ratio_
ind = np.arange(len(ell))
plt. figure(figsize=(15,10))
plt.bar(ind, ell, align='center', alpha=0.5)
plt.plot(np.cumsum(ell))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
```

```python
from pca import pca
# Initialize and keep all PCs
model = pca()
# Fit transform
out = model.fit_transform(X[numeric_columns])
```

```python
out['PC']
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Assuming X is your data
pca = PCA()
Z = pca.fit_transform(X[numeric_columns])

# Check explained variance
explained_variance = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance)

# Set a threshold for cumulative explained variance
threshold = 0.95
num_components = np.argmax(cumulative_explained_variance >= threshold) + 1

# Alternatively, specify the number of components directly
# num_components = 10  # Choose your desired number of components

# Apply PCA with the selected number of components
pca = PCA(n_components=num_components)
Z = pca.fit_transform(X[numeric_columns])
```
